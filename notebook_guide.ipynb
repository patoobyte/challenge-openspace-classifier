## Step 1: Build a Seat

Create a class called `Seat` with two attributes:

- `free` which is a boolean.
- `occupant` which is a string.

and 2 methods (functions): 

- `set_occupant(name)` which allows the program to assign someone a seat if it's free
- `remove_occupant()` which  remove someone from a seat and return the name of the person occupying the seat before


```python
class Seat:
	"""
	Class for a seat.
	"""

	def __init__(self):
		self.free = True
		self.occupant = ""

	def set_occupant(self, occupant):
		if self.free == True:
			self.occupant = occupant
			self.free = False
			print(f"Seat assigned to: {self.occupant}")
	
	def remove_occupant(self):
		if self.free == False:
			print(f"Previous occupant: {self.occupant}")
			self.occupant = ""
			self.free = True
```

## Step 2: Build a Table

Create a class `Table` with ? attributes:

- `capacity` which is an integer
- `seats` which is a list of `Seat` objects (size = `capacity`)

and 3 methods (functions): 
- `has_free_spot()` that returns a boolean (True if a spot is available)
- `assign_seat(name)` that places someone at the table
- `left_capacity()` that returns an integer

Question: Which attributes make sense to give? For now let's say we want to build 6 tables with 4 seats.


```python
class Table:
	"""
	Class for a table, containing seats.
	"""

	def __init__(self, capacity, table_number):
		self.capacity = capacity
		self.seats = [Seat() for _ in range(self.capacity)]
		self.table_number = table_number

	def has_free_spot(self):
		for seat in self.seats:
			if seat.free == True:
				return True
		return False
		
	def assign_seat(self, name):
		for seat in self.seats:
			if seat.free == True:
				seat.set_occupant(name)
				return
		print(f"No available seat at this table for {name}.")
	
	def left_capacity(self):
		left = self.capacity
		for seat in self.seats:
			if seat.free == False:
				left -= 1
		return left
```

```python
# Initialize the table
test_table = Table(4)
# Checks the table capacity
print(test_table.capacity)
# Checks if table has free spot
print(test_table.has_free_spot())
# Assign seats
test_table.assign_seat("Max")
test_table.assign_seat("Ibrahim")
test_table.assign_seat("Irene")
test_table.assign_seat("Imad")
# Assign seat above current capacity
test_table.assign_seat("Soo")
# Checks table's left capacity
print(test_table.left_capacity())
# Checks if table has free spot
print(test_table.has_free_spot())
```
